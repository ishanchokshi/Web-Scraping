# Beautifulsoup starts 

```python
from bs4 import BeautifulSoup
```

```python
page = requests.get(
    "https://www.marketsandmarkets.com/telecom-and-IT-market-research-113.html")
soup = BeautifulSoup(page.content, 'html.parser')

# Extract title of page
page_title = soup.title.text

# print the result
print(page_title)
```

### Extract Body and Head of Webpage

```python
# Extract body of page
page_body = soup.body.txt

# Extract head of page
page_head = soup.head

# print the result
print(page_head)
```

### Printing all h1 tags

```python
# Create all_h1_tags as empty list
all_h1_tags = []

# Set all_h1_tags to all h1 tags of the soup
for element in soup.select('h1'):
    all_h1_tags.append(element.text)



print(all_h1_tags)
```

### Printing all 'a' tags

```python
# Create all_h1_tags as empty list
all_a_tags = []

# Set all_h1_tags to all h1 tags of the soup
for element in soup.select('a'):
    all_a_tags.append(element.text)



print(all_a_tags)
```

## Report Headers

```python
# Create all_h1_tags as empty list
all_h3_tags = []
products = soup.select('tr')
# Set all_h1_tags to all h1 tags of the soup
for element in products:
    
    title = element.select('td > h3 > a')
    
    all_h3_tags.append(title)
report_title = []
all_h3_tags = all_h3_tags[1:51]
for i in range (len(all_h3_tags)):
    tags = str(all_h3_tags[i])
    report_title.append(str(tags.split('">')[1].split('<')[0]))



# print(tags.split('">')[1].split('<')[0])
len(all_h3_tags)
all_h3_tags = all_h3_tags[1:51]
len(report_title)

# tags.split('">')[1].split('<')[0]
```

## Report Links 

```python
links = []

for link in soup.findAll('a'):
    links.append(link.get('href'))
type(links)
```

```python
link = [s for s in links if "Market-Reports" in s]
link
```

## Report Description

```python
# Create all_h1_tags as empty list
all_des_tags = []
products3 = soup.select('tr')
# Set all_h1_tags to all h1 tags of the soup
for element in products3:
    
    title = element.select('td > p')
    
    all_des_tags.append(title)
    
all_des_tags = all_des_tags[1:51]
all_des_tags
report_description = []

for i in range(len(all_des_tags)):
    des_str = str(all_des_tags[i])
    pattern = '<p align="justify">(.*?)</p>'
    report_description.append(re.search(pattern,des_str)[1])



print(report_description[2])
```

### USD and Percent Values
